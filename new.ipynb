```
import numpy as np
import pandas as pd
import random
from collections import defaultdict
```

```
years = [2020 for i in range(20)]
quarters = [i % 4 for i in range(20)]
months = [i%12 for i in range(20)]
bd_name = [random.randint(20000,30000) for i in range(20)]
bd_ids = [random.randint(100,200) for i in range(20)]
balance_type = [random.choice([0,1,2]) for i in range(20)]
```

```
myDict = { k:(v,w) for (k,v,w) in zip(bd_ids, bd_name, balance_type)}

y_q_m = {i:(y,q,m) for i,(y,q,m) in enumerate(zip(years,quarters,months))}
def generate_yqm(period):
  return y_q_m[period]

row = defaultdict(list)
period = 0

for k, (v,w) in myDict.items():
    row[(' ',' ',"BD_ID")].append(k)
    row[(" "," ","BD_NAME")].append(v)
    row[(" "," ","BALANCE_TYPE")].append(w)
    row[("Year","Quarter","Month")].append("")
    y, q, m = generate_yqm(period)
    row[(y,q,m)] = ""
    period+=1
```
